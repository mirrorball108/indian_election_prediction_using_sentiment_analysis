```python
import pandas as pd
import numpy as np
from textblob import TextBlob
import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots
```

```python
modi=pd.read_csv('/content/modi_reviews.csv')
rahul=pd.read_csv('/content/rahul_reviews.csv')
modi.head()
```

```python
rahul.head()
```

```python
modi=modi.iloc[:,1:]
```

```python
rahul=rahul.iloc[:,1:]
```

```python
rahul.head()
```

```python
def findPolarity(text):
  return TextBlob(text).sentiment.polarity
```

```python
modi['Tweet']=modi['Tweet'].astype(str)
rahul['Tweet']=rahul['Tweet'].astype(str)
```

```python
modi['polarity']=modi['Tweet'].apply(findPolarity)
rahul['polarity']=rahul['Tweet'].apply(findPolarity)
```

```python
modi.head()
```

```python
modi.shape
```

```python
rahul.shape
```

```python
rahul.head()
```

```python
def findLabel(polarity):
  if polarity>0:
    return 'Positive'
  elif polarity<0:
    return 'Negative'
  else:
    return 'Neutral'
```

```python
modi['label']=modi['polarity'].apply(findLabel)
rahul['label']=rahul['polarity'].apply(findLabel)
```

```python
rahul.head()
```

```python
modi=modi[modi['label']!='Neutral']
```

```python
rahul[rahul['label']!='Neutral']
```

```python
rahul=rahul[rahul['label']!='Neutral']
```
